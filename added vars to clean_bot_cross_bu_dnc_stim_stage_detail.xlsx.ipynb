```python
import pandas as pd
import matplotlib.pyplot as plt

import datetime

from dateutil.relativedelta import relativedelta

from datetime import date
import numpy as np
import seaborn as sns
import os
import missingno as msno
import warnings

warnings.simplefilter(action = "ignore", category = RuntimeWarning)
```

```python
path = "Clean_Files"
os.makedirs(path, exist_ok = True)

save_path = path + "/"
files = [
"clean_bot_cross_bu_dnc_stim_stage_detail.xlsx",
"clean_gcbu_dnc_recompletions.xlsx",
"clean_gcbu_dnc_stim_stage.xlsx",
"clean_monthly_production.xlsx",
"clean_top_cross_bu_dnc_stim_stage_detail.xlsx"
]
```

```python
df = pd.read_excel(path +files[0])
```

```python
df
```

```python
df = df.filter(['UWI', 'RIG_ACCEPT_DATE', 'RIG_RELEASE_DATE','STG_START_DATE','STG_END_DATE','STG_TOP_DEPTH','STG_BOTTOM_DEPTH','STG_SLURRY_VOLUME_PUMPED'])
df
```

```python
df['STG_END_DATE'] = pd.to_datetime(df['STG_END_DATE'], dayfirst=True)
df['STG_START_DATE'] = pd.to_datetime(df['STG_START_DATE'], dayfirst=True)
df['Stage Time (hrs)'] = df['STG_END_DATE'] - df['STG_START_DATE'] 
df['Stage Time (hrs)']=df['Stage Time (hrs)']/np.timedelta64(1,'h')
```

```python
df['RIG_RELEASE_DATE'] = pd.to_datetime(df['RIG_RELEASE_DATE'], dayfirst=True)
df['RIG_ACCEPT_DATE'] = pd.to_datetime(df['RIG_ACCEPT_DATE'], dayfirst=True)
df['Rig Time (days)'] = (df['RIG_RELEASE_DATE'] - df['RIG_ACCEPT_DATE'])
```

```python
df['Stage Length'] = df['STG_BOTTOM_DEPTH'] - df['STG_TOP_DEPTH'] 
df['Average Stage Depth'] = df[['STG_TOP_DEPTH','STG_BOTTOM_DEPTH']].mean(axis=1)
```

```python
df.to_excel(path + "added_variables" + files[0],index=False)
```
