```python
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams["figure.figsize"] = (20, 4)
import seaborn as sns

import missingno as msno
```

```python
path = "UT_BestFrac_Dataset/"

file_1 = "cross_bu_dnc_stim_stage_detail.xlsx"
file_2 = "cross_bu_dnc_stim_stage_detail.xlsx"
file_3 = "gcbu_dnc_stim_stage.xlsx"
file_4 = "monthly_production.xlsx"
```

```python
df_1 = pd.read_excel(path + file_1)
df_2 = pd.read_excel(path + file_2)
df_3 = pd.read_excel(path + file_3)
df_4= pd.read_excel(path + file_4)
```

```python
msno.matrix(df_1)
```

```python
df_1_empty = pd.DataFrame(columns=["Feature", "% Complete"])

nans_1 = df_1.isnull().sum(axis = 0).tolist()
for feature in range(0, len(df_1.columns)):
    df_1_empty.loc[feature, "Feature"] = df_1.columns[feature]
    df_1_empty.loc[feature, "% Complete"] = 1.0 - nans_1[feature] / len(df_1.index)

df_1_empty
```
