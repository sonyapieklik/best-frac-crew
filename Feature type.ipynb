```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import math

import missingno as msno
```

```python
save_name = "Clean_Files"
os.makedirs(save_name, exist_ok = True)

save_path = save_name + "/"
```

```python
original_path = "UT_BestFrac_Dataset/"

files = [
    "cross_bu_dnc_stim_stage_detail.xlsx",
    "gcbu_dnc_recompletions.xlsx",
    "gcbu_dnc_stim_stage.xlsx",
    "monthly_production.xlsx"
]
```

```python
df_1 = pd.read_excel(original_path + files[0])
df_2 = pd.read_excel(original_path + files[1])
df_3 = pd.read_excel(original_path + files[2])
df_4 = pd.read_excel(original_path + files[3])
```

```python
df_1_features = pd.DataFrame(columns=["Feature", "Type"])
out_index = 0
for feature in df_1.columns:
    index = 0
    df_1_features.loc[out_index, "Feature"] = df_1.columns[out_index]
    while(index < len(df_1)):
        cur = df_1.loc[index, df_1.columns[out_index]]
        if(type(cur) == type("")):
            cur = "STRING"
            break
        if(not np.isnan(cur)):
            cur = "INT"
            break
        index += 1
    
    if(index != len(df_1)):
        df_1_features.loc[out_index, "Type"] = cur
    else:
        df_1_features.loc[out_index, "Type"] = "NAN"
    print("Feature: %s Type: %s" % (df_1_features.loc[out_index, "Feature"], df_1_features.loc[out_index, "Type"]))
    out_index += 1
```
