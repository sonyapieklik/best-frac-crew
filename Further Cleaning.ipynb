```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import os
import math

import missingno as msno
```

```python
clean_name = "Clean_Files"
clean_path = clean_name + "/clean_"

files = [
    "top_cross_bu_dnc_stim_stage_detail.xlsx",
    "bot_cross_bu_dnc_stim_stage_detail.xlsx",
    "gcbu_dnc_recompletions.xlsx",
    "gcbu_dnc_stim_stage.xlsx",
    "monthly_production.xlsx"
]
```

```python
df_1_a = pd.read_excel(clean_path + files[0])
df_1_b = pd.read_excel(clean_path + files[1])
df_2 = pd.read_excel(clean_path + files[2])
df_3 = pd.read_excel(clean_path + files[3])
df_4 = pd.read_excel(clean_path + files[4])
```

```python
df_1_types = pd.read_excel("Clean_Files/DF_1_Feature_Types.xlsx")
df_2_types = pd.read_excel("Clean_Files/DF_2_Feature_Types.xlsx")
df_3_types = pd.read_excel("Clean_Files/DF_3_Feature_Types.xlsx")
df_4_types = pd.read_excel("Clean_Files/DF_4_Feature_Types.xlsx")
```

## File 1, top part

```python
msno.matrix(df_1_a)
```

```python
cur_df = df_1_a
comp_df = df_1_types

to_drop = []

for feature in cur_df.columns:
    temp_df = comp_df.loc[comp_df["Feature"] == feature].reset_index()
    if(len(temp_df) != 0 and temp_df.loc[0, "Type"] == "STRING"):
        to_drop.append(feature)
        
cur_df = cur_df.drop(columns = to_drop)
msno.matrix(cur_df)
cur_df.to_excel(clean_name + "/only_Int_" + files[0])
```
